```python
from sklearn.preprocessing import LabelEncoder
def dummyEncode(df):
        columnsToEncode = list(df.select_dtypes(include=['category','object']))
        le = LabelEncoder()
        for feature in columnsToEncode:
            try:
                df[feature] = le.fit_transform(df[feature])
            except:
                print('Error encoding '+feature)
        return df
```

```python
#Replace NaN values with 0
df['duration'] = df['duration'].str.replace('-','0')
df['orig_bytes'] = df['orig_bytes'].str.replace('-','0')
df['resp_bytes'] = df['resp_bytes'].str.replace('-','0')
```

```python
#convert the following columns from float64 to int32
df['missed_bytes'] = df['missed_bytes'].astype('int')
df['id.orig_p'] = df['id.orig_p'].astype('int')
df['id.resp_p'] = df['id.resp_p'].astype('int')

df['orig_pkts'] = df['orig_pkts'].astype('int')
df['resp_pkts'] = df['resp_pkts'].astype('int')
```

```python
#Assign down sampling ratios for required samples
sampling_strategy = {6: 309637, 5: 190141, 0: 136642, 3: 81340}
```

```python
#split df into features and labels
df_labels = df['label']
df_features = df.drop(columns=['label'])
```

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy=sampling_strategy)
```

```python
#before undersampling value count
df['label'].value_counts()
```

```python
df_features, df_labels = undersample.fit_resample(df_features, df_labels)
```

```python
#after undersampling value count
df_labels.value_counts()
```

```python
df_features.insert(12, 'label', df_labels)
df = df_features
```
